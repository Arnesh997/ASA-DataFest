```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Analysing items.csv

```python
item_path = "../Data/items.csv"
item_pd = pd.read_csv(item_path)
print("Number of items: ", len(item_pd))
```

```python
unique_institutions = item_pd["institution_id"].unique()
print("Number of unique institutions: ", len(unique_institutions))
print("Institution IDs: ", unique_institutions)
```

```python
# Counting the number of rows with a specific institution_id
for inst in unique_institutions:
    print("Number of items with institution_id ", inst, ": ", len(item_pd[item_pd["institution_id"] == inst]))
```

```python
# Removing rows with institution_ids 94a809a9-a0ef-4c47-8d96-3a5ad76f674b and 364da48a-e0b2-4507-bc31-e7761fe16e95
item_pd = item_pd.drop(item_pd[item_pd["institution_id"] == "94a809a9-a0ef-4c47-8d96-3a5ad76f674b"].index)
item_pd = item_pd.drop(item_pd[item_pd["institution_id"] == "364da48a-e0b2-4507-bc31-e7761fe16e95"].index)
```

```python
# Removing the following columns (Since variable_list.csv says that they are mostly/All empty)
# Columns: dcl_hint, dcl_pre_exercise_code, dcl_sample_code, dcl_sct
# Notes about columns: dcl_solution is removed since all NA values
item_pd = item_pd.drop(columns=["dcl_hint", "dcl_pre_exercise_code", "dcl_sample_code", "dcl_sct", "dcl_solution"])
item_pd
```

Now, I am looking into only one specific institution. That institution seems to have 2 classes

```python
unique_class = item_pd["class_id"].unique()
print("Number of unique classes: ", len(unique_class))
print("Class IDs: ", unique_class)
# Counting the number of rows with a specific institution_id
for cls in unique_class:
    print("Number of items with class ID ", cls, ": ", len(item_pd[item_pd["class_id"] == cls]))
item_pd
```
