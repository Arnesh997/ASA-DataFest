```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Analysing items.csv

```python
item_path = "../Data/items.csv"
item_pd = pd.read_csv(item_path)
print("Number of items: ", len(item_pd))
```

```python
unique_institutions = item_pd["institution_id"].unique()
print("Number of unique institutions: ", len(unique_institutions))
print("Institution IDs: ", unique_institutions)
```

```python
# Counting the number of rows with a specific institution_id
for inst in unique_institutions:
    print("Number of items with institution_id ", inst, ": ", len(item_pd[item_pd["institution_id"] == inst]))
```

```python
# Removing rows with institution_ids 94a809a9-a0ef-4c47-8d96-3a5ad76f674b and 364da48a-e0b2-4507-bc31-e7761fe16e95
item_pd = item_pd.drop(item_pd[item_pd["institution_id"] == "94a809a9-a0ef-4c47-8d96-3a5ad76f674b"].index)
item_pd = item_pd.drop(item_pd[item_pd["institution_id"] == "364da48a-e0b2-4507-bc31-e7761fe16e95"].index)
```

```python
# Removing the following columns (Since variable_list.csv says that they are mostly/All empty)
# Columns: dcl_hint, dcl_pre_exercise_code, dcl_sample_code, dcl_sct
# Notes about columns: dcl_solution is removed since all NA values
item_pd = item_pd.drop(columns=["dcl_hint", "dcl_pre_exercise_code", "dcl_sample_code", "dcl_sct", "dcl_solution"])
item_pd
```

Now, I am looking into only one specific institution. That institution seems to have 2 classes

```python
unique_class = item_pd["class_id"].unique()
print("Number of unique classes: ", len(unique_class))
print("Class IDs: ", unique_class)
# Counting the number of rows with a specific institution_id
for cls in unique_class:
    print("Number of items with class ID ", cls, ": ", len(item_pd[item_pd["class_id"] == cls]))
item_pd
```

```python
uniqueQref = item_pd["lrn_question_reference"].unique()
len(uniqueQref)
# Each of the questions have a unique lrn_question_reference --> Dropping the column
item_pd = item_pd.drop(columns=["lrn_question_reference"])

# Number of NaN in lrn_activity_reference
print("Number of NaN in lrn_activity_reference: ", item_pd["lrn_activity_reference"].isna().sum())
# Non-NaN values in lrn_activity_reference
print("Number of non-NaN in lrn_activity_reference: ", item_pd["lrn_activity_reference"].notna().sum())
# Printing the unique values in lrn_activity_reference
uniqueActRef = item_pd["lrn_activity_reference"].unique()
print("Number of unique lrn_activity_reference: ", len(uniqueActRef))
print("Unique lrn_activity_reference: ", uniqueActRef)
```

```python
# All thr unique values in lrn_type column
uniqueLrnType = item_pd["lrn_type"].unique()
print("Number of unique lrn_type: ", len(uniqueLrnType))
print("Unique lrn_type: ", uniqueLrnType)  
# Number of NaA in lrn_type
print("Number of NaN in lrn_type: ", item_pd["lrn_type"].isna().sum())
```

```python
# Number of True and False in review_flag column
print("Number of TRUE in review_flag: ", len(item_pd[item_pd["review_flag"] == True]))
print("Number of FALSE in review_flag: ", len(item_pd[item_pd["review_flag"] == False]))
# Plotting a graph to understand the type of lrn_type questions that was part of chapter review (review_flag = True)
# On top of each bar, the number of review_flag = True is displayed
# Graph 1
review_flag = item_pd[item_pd["review_flag"] == True]
review_flag = review_flag["lrn_type"].value_counts()
review_flag.plot(kind="bar")
for i in range(len(review_flag)):
    plt.text(i, review_flag[i], review_flag[i], ha="center")
plt.title("Number of review_flag = True for each lrn_type")
plt.show()
# Graph 2
plt.figure()
item_pd.groupby("lrn_type")["review_flag"].value_counts().unstack().plot(kind='bar')
plt.title("Number of review_flag TRUE and FALSE for each lrn_type")
plt.show()
```

```python
# Understanding the column 'item_type' by plotting different graphs with other columns
# Graph 1
item_type = item_pd["item_type"].value_counts()
item_type.plot(kind="bar")
for i in range(len(item_type)):
    plt.text(i, item_type[i], item_type[i], ha="center")
plt.title("Number of items for each item_type")
plt.show()
# Graph 2
plt.figure()
item_pd.groupby("item_type")["lrn_type"].value_counts().unstack().plot(kind='bar')
plt.title("Number of lrn_type for each item_type")
plt.show()
# Graph 3
plt.figure()
item_pd.groupby("item_type")["review_flag"].value_counts().unstack().plot(kind='bar')
plt.title("Number of review_flag for each item_type")
plt.show()
# Graph 4
plt.figure()
item_pd.groupby("item_type")["institution_id"].value_counts().unstack().plot(kind='bar')
plt.title("Number of institution_id for each item_type")
plt.show()
```

# Insights and Analysis

## Item Types Distribution
- **Most Common Types**: The data shows a predominance of "learnosity" (Learnosity questions) and "mcq" (multiple choice questions), indicating these are the most frequently used question types.
- **Code Items**: There's a noticeable scarcity of "code" items. This suggests that the course leans more towards enhancing conceptual understanding rather than coding skills.

## Organization of Content
- **Question Clustering**: Questions often appear grouped together by chapter, section, and page (`chapter_number`, `section_number`, and `page` columns). This clustering implies a structured approach to content organization.

## Question Formats
- **Diversity in Formats**: The `lrn_template_name` column showcases a variety of question formats including multiple choice, short text, essay, and matching. This diversity is likely aimed at engaging learners through different types of interactions.

## Review Questions
- **Review Flag Usage**: Boolean values in the `review_flag` column indicate that some questions are specifically marked for review. The initial chapters (1-4) host the majority of these review-oriented questions.

## Learnosity Integration
- **Learnosity Items**: Items tagged as "learnosity" and "learnosity-activity" in the `item_type` column are often associated with `lrn_activity_reference` values, suggesting integration with Learnosity activities.
- **Question Positioning**: The `lrn_question_position` column details the sequence of sub-questions within an item, revealing a trend towards multipart questions, particularly from chapter 4 onwards.

## Recurring Assessments
- **Pulse Checks**: Identifiers like "Pulse_a2" recurring across chapters indicate the use of periodic "pulse check" assessments to gauge learning progression.

## Evolving Complexity
- **New Question Types**: Introduction of new `lrn_template_names` such as "formulaV2" and "choicematrix" in chapter 5 points towards the inclusion of more complex question types as learners advance in the course.
